## **Fraudulent Claim Detection**

## Problem Statement
Global Insure, a leading insurance company, processes thousands of claims annually. However, a significant percentage of these claims turn out to be fraudulent, resulting in considerable financial losses. The company’s current process for identifying fraudulent claims involves manual inspections, which is time-consuming and inefficient. Fraudulent claims are often detected too late in the process, after the company has already paid out significant amounts. Global Insure wants to improve its fraud detection process using data-driven insights to classify claims as fraudulent or legitimate early in the approval process. This would minimise financial losses and optimise the overall claims handling process.

## Business Objective
Global Insure wants to build a model to classify insurance claims as either fraudulent or legitimate based on historical claim details and customer profiles. By using features like claim amounts, customer profiles and claim types, the company aims to predict which claims are likely to be fraudulent before they are approved.

#### **2.2.2** Identify and drop any columns that are completely empty <font color="red">[1 Mark]</font>

#### **2.2.3** Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values <font color="red">[1 Mark]</font>

```python
# Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Check for negative values in numeric columns only
negative_values = (df[numeric_cols] < 0).sum()
print("Negative values in the dataset:")
print(negative_values[negative_values > 0])

# Look at rows with negative values
negative_rows = df[df[numeric_cols].lt(0).any(axis=1)]
print(f"\nNumber of rows with negative values: {len(negative_rows)}")

# Drop rows with negative values in columns that should only have positive values
# (excluding 'capital-loss' which can be negative)
print("Dropping rows with negative values in numeric columns (excluding 'capital-loss')")
cols_to_check = [col for col in numeric_cols if col != 'capital-loss']
df = df[~df[cols_to_check].lt(0).any(axis=1)]
print(f"Dataset shape after removing rows with negative values: {df.shape}")
```

#### **2.2.4** Identify and remove columns where a large proportion of the values are unique or near-unique, as these columns are likely to be identifiers or have very limited predictive power <font color="red">[1 Mark]</font>

```python
# Identify and remove columns that are likely to be identifiers or have very limited predictive power
# Calculate the percentage of unique values in each column
unique_counts = {col: df[col].nunique() / len(df) for col in df.columns}

# Sort columns by the percentage of unique values (highest to lowest)
sorted_unique_counts = sorted(unique_counts.items(), key=lambda x: x[1], reverse=True)

# Display columns with their percentage of unique values
print("Columns with their percentage of unique values:")
for col, ratio in sorted_unique_counts:
    print(f"{col}: {ratio:.4f} ({df[col].nunique()} / {len(df)})")

# Identify columns with high cardinality (e.g., more than 80% unique values)
high_cardinality_cols = [col for col, ratio in unique_counts.items() if ratio > 0.8]
print("\nColumns with high cardinality (>80% unique values):")
print(high_cardinality_cols)

# Remove columns with high cardinality 
if high_cardinality_cols:
    print(f"\nRemoving {len(high_cardinality_cols)} columns with high cardinality")
    df.drop(columns=high_cardinality_cols, inplace=True)
    print(f"Dataset shape after removing high cardinality columns: {df.shape}")
else:
    print("\nNo high cardinality columns found to remove")
```

#### **4.1.2** Visualise the distribution of selected numerical features using appropriate plots to understand their characteristics <font color = "red">[4 Marks]</font>

```python
# Plot all the numerical columns to understand their distribution
# Layout setup
n_cols = 3
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
plt.figure(figsize=(n_cols * 5.5, n_rows * 4.5))
sns.set(style="whitegrid")

# Loop through each numeric column
for i, col in enumerate(numerical_cols, 1):
    # Compute statistics
    mean_val = X_train[col].mean()
    median_val = X_train[col].median()
    skew_val = X_train[col].skew()
    
    # Determine shape
    if abs(skew_val) < 0.5:
        shape = "Symmetric"
    elif skew_val > 0.5:
        shape = "Right-skewed"
    else:
        shape = "Left-skewed"
    
    # Plot histogram with KDE
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(X_train[col], kde=True, bins=30, color='skyblue')
    
    # Dynamic title with stats & interpretation
    plt.title(f'{col}\nMean: {mean_val:.2f}, Median: {median_val:.2f}, Skew: {skew_val:.2f}\n{shape}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create boxplots to identify outliers in numerical features
plt.figure(figsize=(n_cols * 6.5, n_rows * 5.5))
sns.set(style="whitegrid", palette="pastel", font_scale=1.1)

# Plot boxplots for each numerical column
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot( x=X_train[col],
        color=sns.color_palette("pastel")[i % len(sns.color_palette("pastel"))],
        width=0.5,
        fliersize=3,  # smaller outlier points
        linewidth=1.5)
    
    # Calculate IQR for outlier detection
    Q1 = X_train[col].quantile(0.25)
    Q3 = X_train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = X_train[(X_train[col] < lower_bound) | (X_train[col] > upper_bound)]
    
    # Annotated title
    plt.title(f"{col}\nOutliers: {len(outliers)} | Range: ({lower_bound:.1f}, {upper_bound:.1f})",
             fontsize=11,
        fontweight='bold')
    plt.xlabel(col, fontsize=10)
    plt.grid(True, axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.suptitle("Boxplots with Outlier Summary", fontsize=18, y=1.02)
plt.show()

# Print detailed outlier observations
print("\nObservations from Boxplot Plots:")
for col in numerical_cols:
    Q1, Q3 = X_train[col].quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = X_train[(X_train[col] < lower) | (X_train[col] > upper)]

    print(f"{col:<25} | {len(outliers):<4} | Range: ({lower:.2f}, {upper:.2f})")
```

### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features

# Define function to calculate and visualize target likelihood for categorical features
def analyze_categorical_target_likelihood(X_df, y_series, top_n=10):
    """
    Analyze the target event likelihood for each level of categorical features
    
    Parameters:
    -----------
    X_df : pandas DataFrame
        DataFrame containing features
    y_series : pandas Series
        Target variable (binary)
    top_n : int
        Show only top_n categories for features with many levels
    
    Returns:
    --------
    dict : Dictionary containing variance of fraud rates for each categorical feature
    """
    # Combine features and target for analysis
    data = X_df.copy()
    data['fraud_reported'] = y_series
    
    # Identify categorical columns
    cat_cols = X_df.select_dtypes(include=['object']).columns.tolist()
    print(f"Found {len(cat_cols)} categorical features")
    
    # Dictionary to store feature importance based on variance in fraud rates
    feature_importance = {}
    
    # Plot settings
    n_cols = 2
    n_rows = (len(cat_cols) + n_cols - 1) // n_cols  # Ceiling division
    plt.figure(figsize=(18, 5 * n_rows))
    
    # Calculate overall average fraud rate once
    avg_fraud_rate = data[data['fraud_reported'] == 'Y'].shape[0] / data.shape[0] * 100
    
    # Analyze each categorical feature
    for i, col in enumerate(cat_cols):
        # Count total occurrences and fraud occurrences for each category
        category_counts = data.groupby(col).size()
        fraud_counts = data[data['fraud_reported'] == 'Y'].groupby(col).size()
        
        # Calculate fraud rates (% of fraudulent claims in each category)
        fraud_rate = pd.DataFrame({
            'total_count': category_counts,
            'fraud_count': fraud_counts.reindex(category_counts.index).fillna(0),
        })
        fraud_rate['fraud_rate'] = fraud_rate['fraud_count'] / fraud_rate['total_count'] * 100
        fraud_rate['non_fraud_rate'] = 100 - fraud_rate['fraud_rate']
        
        # Sort by fraud rate (descending)
        fraud_rate = fraud_rate.sort_values('fraud_rate', ascending=False)
        
        # Limit display to top_n categories if there are too many
        if len(fraud_rate) > top_n:
            display_data = fraud_rate.head(top_n)
            too_many = True
        else:
            display_data = fraud_rate
            too_many = False
        
        # Calculate the variance of fraud rates (used to measure feature importance)
        # Higher variance means the feature distinguishes better between fraud and non-fraud
        feature_importance[col] = fraud_rate['fraud_rate'].var()
        
        # Plot the fraud rates
        ax = plt.subplot(n_rows, n_cols, i+1)
        
        # Create stacked bar chart
        bars = plt.bar(display_data.index, display_data['fraud_rate'], color='red', label='Fraud')
        plt.bar(display_data.index, display_data['non_fraud_rate'], bottom=display_data['fraud_rate'], 
                color='green', label='Non-Fraud')
        
        # Add percentage labels on bars
        for bar, fraud_pct in zip(bars, display_data['fraud_rate']):
            height = bar.get_height()
            if height > 5:  # Only show percentage if it's at least 5%
                plt.text(bar.get_x() + bar.get_width()/2., height/2,
                         f'{fraud_pct:.1f}%',
                         ha='center', va='center', color='white', fontweight='bold')
        
        # Add plot info with index in the title
        plt.title(f'Plot {i+1}: Fraud Rate by {col}', fontsize=12)
        if too_many:
            plt.xlabel(f'Top {top_n} categories (out of {len(fraud_rate)})', fontsize=10)
        else:
            plt.xlabel(col, fontsize=10)
        plt.ylabel('Percentage', fontsize=10)
        plt.xticks(rotation=45, ha='right')
        plt.grid(axis='y', alpha=0.3)
        
        # Add legend for first plot
        if i == 0:
            plt.legend(loc='best')



    plt.tight_layout()
    plt.show()
    
    # Rank features by variance in fraud rates
    feature_importance = pd.Series(feature_importance).sort_values(ascending=False)
    
    print("\nFeature importance based on variance in fraud rates:")
    for feature, variance in feature_importance.items():
        print(f"{feature}: {variance:.4f}")
    
    print("\nCategorical features with low variance may not contribute much to explaining fraud.")
    return feature_importance

# Run the analysis on training data
cat_feature_importance = analyze_categorical_target_likelihood(X_train, y_train)
```

### **6.2 Feature Creation** <font color="red">[4 marks]</font>

Create new features from existing ones to enhance the model's ability to capture patterns in the data. This may involve deriving features from date/time columns, combining features, or creating interaction terms.

```python
# Create copies to avoid DataFrame SettingWithCopyWarning
X_train = X_train.copy()
X_test = X_test.copy()

import numpy as np

# 1. Extract date-based features from incident_date
print("Creating date-based features...")
# Extract day of week (0: Monday, 6: Sunday)
X_train['incident_day_of_week'] = X_train['incident_date'].dt.dayofweek
X_test['incident_day_of_week'] = X_test['incident_date'].dt.dayofweek

# Extract month
X_train['incident_month'] = X_train['incident_date'].dt.month
X_test['incident_month'] = X_test['incident_date'].dt.month

# Extract if weekend (5,6 are weekend days)
X_train['is_weekend'] = X_train['incident_day_of_week'].isin([5, 6]).astype(int)
X_test['is_weekend'] = X_test['incident_day_of_week'].isin([5, 6]).astype(int)

# 2. Create vehicle age feature
current_year = 2015  # Based on incident_date in dataset which is 2015
X_train['vehicle_age'] = current_year - X_train['auto_year']
X_test['vehicle_age'] = current_year - X_test['auto_year']

# 3. Create claim amount ratio features
print("Creating claim ratio features...")
# Injury to total claim ratio (with zero handling)
X_train['injury_claim_ratio'] = np.where(X_train['total_claim_amount'] > 0, 
                                        X_train['injury_claim'] / X_train['total_claim_amount'], 0)
X_test['injury_claim_ratio'] = np.where(X_test['total_claim_amount'] > 0, 
                                      X_test['injury_claim'] / X_test['total_claim_amount'], 0)

# Property to total claim ratio
X_train['property_claim_ratio'] = np.where(X_train['total_claim_amount'] > 0, 
                                          X_train['property_claim'] / X_train['total_claim_amount'], 0)
X_test['property_claim_ratio'] = np.where(X_test['total_claim_amount'] > 0, 
                                        X_test['property_claim'] / X_test['total_claim_amount'], 0)

# Vehicle to total claim ratio
X_train['vehicle_claim_ratio'] = np.where(X_train['total_claim_amount'] > 0, 
                                         X_train['vehicle_claim'] / X_train['total_claim_amount'], 0)
X_test['vehicle_claim_ratio'] = np.where(X_test['total_claim_amount'] > 0, 
                                       X_test['vehicle_claim'] / X_test['total_claim_amount'], 0)

# 4. Create hour-based features
print("Creating time-of-day features...")
def categorize_hour(hour):
    if 5 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 20:
        return 'Evening'
    else:
        return 'Night'

X_train['incident_time_of_day'] = X_train['incident_hour_of_the_day'].apply(categorize_hour)
X_test['incident_time_of_day'] = X_test['incident_hour_of_the_day'].apply(categorize_hour)

# 5. Create interaction features between important numerical variables
print("Creating interaction features...")
# Interaction between vehicles involved and witnesses
X_train['vehicles_witnesses_interaction'] = X_train['number_of_vehicles_involved'] * X_train['witnesses']
X_test['vehicles_witnesses_interaction'] = X_test['number_of_vehicles_involved'] * X_test['witnesses']

# Interaction between bodily injuries and vehicles involved
X_train['injuries_vehicles_interaction'] = X_train['bodily_injuries'] * X_train['number_of_vehicles_involved']
X_test['injuries_vehicles_interaction'] = X_test['number_of_vehicles_involved'] * X_test['bodily_injuries']

# 6. Create age groups for easier pattern recognition
print("Creating age group features...")
def categorize_age(age):
    if age < 25:
        return 'Young'
    elif 25 <= age < 40:
        return 'Adult'
    elif 40 <= age < 60:
        return 'Middle-aged'
    else:
        return 'Senior'

X_train['age_group'] = X_train['age'].apply(categorize_age)
X_test['age_group'] = X_test['age'].apply(categorize_age)

# 7. Create customer tenure groups
print("Creating customer tenure features...")
def categorize_tenure(months):
    if months < 50:
        return 'New'
    elif 50 <= months < 200:
        return 'Regular'
    else:
        return 'Loyal'

X_train['customer_tenure_group'] = X_train['months_as_customer'].apply(categorize_tenure)
X_test['customer_tenure_group'] = X_test['months_as_customer'].apply(categorize_tenure)

# 8. Create a high claim amount flag
high_claim_threshold = X_train['total_claim_amount'].quantile(0.75)
X_train['high_claim_amount'] = (X_train['total_claim_amount'] > high_claim_threshold).astype(int)
X_test['high_claim_amount'] = (X_test['total_claim_amount'] > high_claim_threshold).astype(int)

# Display the new features
print("\nNew features created successfully!")
print(f"Training set shape after feature creation: {X_train.shape}")
print(f"Test set shape after feature creation: {X_test.shape}")
print("\nNew features:")
new_features = [col for col in X_train.columns if col not in numerical_cols and col not in X_train.select_dtypes(include=['object']).columns 
               and col != 'incident_date']
print(new_features)


```

### **6.3 Handle redundant columns** <font color="red">[3 marks]</font>

Analyse the data to identify features that may be redundant or contribute minimal information toward predicting the target variable and drop them.

- You can consider features that exhibit high correlation with other variables, which you may have observed during the EDA phase.
- Features that don't strongly influence the prediction, which you may have observed during the EDA phase.
- Categorical columns with low value counts for some levels can be remapped to reduce number of unique levels, and features with very high counts for just one level may be removed, as they resemble unique identifier columns and do not provide substantial predictive value.
- Additionally, eliminate any columns from which the necessary features have already been extracted in the preceding step.

```python
# Drop redundant columns from training and validation data
# 1. Check for highly correlated numerical features
print("Checking for highly correlated features...")
corr_matrix = X_train.select_dtypes(include=['int64', 'float64']).corr()

# Set a correlation threshold to identify highly correlated pairs
threshold = 0.85
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col_i = corr_matrix.columns[i]
            col_j = corr_matrix.columns[j]
            high_corr_pairs.append((col_i, col_j, corr_matrix.iloc[i, j]))

print(f"Found {len(high_corr_pairs)} highly correlated pairs (correlation > {threshold}):")
for col1, col2, corr in high_corr_pairs:
    print(f"  - {col1} & {col2}: {corr:.4f}")

# 2. Identify columns to drop
columns_to_drop = []

# Drop columns from which we've already extracted features
derived_features = {
    'incident_date': ['incident_day_of_week', 'incident_month', 'is_weekend'],
    'auto_year': ['vehicle_age'],
    'incident_hour_of_the_day': ['incident_time_of_day']
}

for source_col, derived_cols in derived_features.items():
    # Check if source column exists and all derived features exist
    if source_col in X_train.columns and all(col in X_train.columns for col in derived_cols):
        columns_to_drop.append(source_col)
        print(f"\nDropping '{source_col}' as we've created derived features from it: {', '.join(derived_cols)}")

# 3. Drop columns with high multicollinearity
claim_cols_corr = {}
if 'total_claim_amount' in X_train.columns:
    for col in ['vehicle_claim', 'property_claim', 'injury_claim']:
        if col in X_train.columns:
            corr = X_train[['total_claim_amount', col]].corr().iloc[0, 1]
            claim_cols_corr[col] = corr
            if corr > 0.9:
                columns_to_drop.append(col)
                print(f"Dropping '{col}' due to high correlation ({corr:.4f}) with 'total_claim_amount'")

# 4. Handle highly correlated features like months_as_customer and age
if 'months_as_customer' in X_train.columns and 'age' in X_train.columns:
    corr = X_train[['months_as_customer', 'age']].corr().iloc[0, 1]
    if corr > 0.9:
        columns_to_drop.append('months_as_customer')
        print(f"Dropping 'months_as_customer' due to high correlation ({corr:.4f}) with 'age'")

# 5. Remove categorical columns with too many or too few unique values
cat_cols = X_train.select_dtypes(include=['object']).columns
for col in cat_cols:
    # If a categorical column has only one unique value, drop it
    if X_train[col].nunique() == 1:
        columns_to_drop.append(col)
        print(f"Dropping '{col}' as it has only one unique value")
    # If a categorical column has too many unique values (like an identifier), consider dropping it
    elif X_train[col].nunique() / len(X_train) > 0.8:
        columns_to_drop.append(col)
        print(f"Dropping '{col}' as it has too many unique values ({X_train[col].nunique()} values, likely an identifier)")

# Finally drop all identified columns (avoiding duplicates)
columns_to_drop = list(set(columns_to_drop))  # Remove duplicates
if columns_to_drop:
    X_train = X_train.drop(columns=columns_to_drop)
    X_test = X_test.drop(columns=[col for col in columns_to_drop if col in X_test.columns])
    print(f"\nRemoved {len(columns_to_drop)} redundant columns: {columns_to_drop}")
    print(f"Training data shape after removing redundant columns: {X_train.shape}")
    print(f"Testing data shape after removing redundant columns: {X_test.shape}")
else:
    print("\nNo redundant columns found to remove.")
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
# Identify categorical columns (excluding target)
cat_cols = X_train.select_dtypes(include=['object']).columns
print(f"Found {len(cat_cols)} categorical features to analyze for combining values")

# Dictionary to store value mappings for each categorical column
category_mappings = {}

# Function to check if a value should be treated as missing/unknown
def is_unknown(val):
    return val in ['?', 'None', None, np.nan]

# Loop through each categorical column
for col in cat_cols:
    # Get value counts in training data
    value_counts = X_train[col].value_counts()
    total_count = len(X_train)
    
    # Define threshold for rare categories (less than 2% of data)
    rare_threshold = total_count * 0.02
    
    # Identify rare categories
    rare_categories = value_counts[value_counts < rare_threshold].index.tolist()
    
    # Print details about the column
    print(f"\nColumn '{col}'")
    print(f"  Total unique values: {X_train[col].nunique()}")
    print(f"  Rare categories (< 2% of data): {len(rare_categories)}")
    
    if rare_categories:
        # Create mapping for this column
        mapping = {category: 'Other' if category in rare_categories else category 
                  for category in X_train[col].unique()}
        
        # Handle missing/unknown values consistently
        for unknown_val in ['?', 'None', None, np.nan]:
            mapping[unknown_val] = 'Unknown'
            
        category_mappings[col] = mapping
        
        # Apply mapping to train data
        X_train[col] = X_train[col].map(lambda x: 'Unknown' if is_unknown(x) else mapping.get(x, 'Other'))
        
        # Apply mapping to test data, handling any new categories in test
        X_test[col] = X_test[col].map(lambda x: 'Unknown' if is_unknown(x) else mapping.get(x, 'Other'))
        
        # Count before and after to show reduction in categories
        before_count = len(value_counts)
        after_count = X_train[col].nunique()
        print(f"  Reduced categories from {before_count} to {after_count}")
        print(f"  Top 5 categories after combining: {X_train[col].value_counts().nlargest(5).to_dict()}")
    else:
        print(f"  No rare categories to combine")
        
        # Still handle missing/unknown values consistently even if no rare categories
        X_train[col] = X_train[col].map(lambda x: 'Unknown' if is_unknown(x) else x)
        X_test[col] = X_test[col].map(lambda x: 'Unknown' if is_unknown(x) else x)

# Special handling for specific columns
special_columns = [col for col in cat_cols if col in ['authorities_contacted', 'collision_type', 'property_damage', 'police_report_available']]
for col in special_columns:
    if col in X_train.columns:
        # Check for nulls after our processing
        null_count_train = X_train[col].isnull().sum()
        null_count_test = X_test[col].isnull().sum()
        
        if null_count_train > 0 or null_count_test > 0:
            X_train[col] = X_train[col].fillna('Unknown')
            X_test[col] = X_test[col].fillna('Unknown')
            print(f"Fixed remaining {null_count_train + null_count_test} null values in '{col}'")

# Print summary of transformations
columns_modified = len(category_mappings)
print(f"\nModified {columns_modified} categorical columns by combining rare categories")

# Print detailed info about the distribution in key categorical features
important_cols = ['incident_severity', 'insured_hobbies', 'incident_type', 'auto_make', 'insured_relationship']
for col in important_cols:
    if col in X_train.columns:
        print(f"\n{col} value distribution:")
        dist = X_train[col].value_counts(normalize=True) * 100
        for val, pct in dist.items():
            print(f"  {val}: {pct:.1f}%")
```

### **7.1 Feature selection** <font color = red>[4 marks]</font>
Identify and select the most relevant features for building a logistic regression model using Recursive Feature Elimination with Cross-Validation (RFECV).

#### **7.1.1** Import necessary libraries <font color="red">[1 Mark]</font>

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np
```

#### **7.1.2** Perform feature selection <font color="red">[2 Mark]</font>

```python
# Apply RFECV to identify the most relevant features

# Use a basic Logistic Regression model for feature selection
# Use StratifiedKFold for cross-validation to maintain class balance

# Initialize the Logistic Regression model with balanced class weights
logreg = LogisticRegression(
    solver='liblinear',  # Good for small datasets
    class_weight='balanced',  # Account for class imbalance
    max_iter=1000,  # Increase iterations for convergence
    random_state=42
)

# Set up cross-validation with StratifiedKFold to maintain class proportions
cv_stratified = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize RFECV with the logistic regression model
rfecv = RFECV(
    estimator=logreg,
    step=1,  # Remove one feature at a time
    cv=cv_stratified,  #  Use stratified cross-validation
    scoring='roc_auc',  # Use AUC-ROC as the evaluation metric
    min_features_to_select=5,  # At least 5 features should remain
    n_jobs=-1,  # Use all available cores
    verbose=1  # Show progress
)

# Fit RFECV to the training data
print("Starting recursive feature elimination with cross-validation...")
rfecv.fit(X_train, y_train)

# Get the optimal number of features
n_features_optimal = rfecv.n_features_
print(f"\nOptimal number of features: {n_features_optimal}")

# Get the selected features
selected_features = X_train.columns[rfecv.support_]
print(f"\nSelected {len(selected_features)} features:")
print(selected_features[:n_features_optimal])  # 

```

Evaluating these metrics is crucial for detecting multicollinearity and ensuring that the selected predictors are not highly correlated.

```python
# Select only the columns selected by RFECV
X_train_logit = X_train[col]
```

#### **7.2.2** Fit logistic regression model <font color="red">[2 Marks]</font>

```python
# Fit a logistic Regression model on X_train after adding a constant and output the summary
# Create a logistic regression model using statsmodels
logit_model = sm.Logit(y_train, X_train_logit_const)

# Fit the model using maximum likelihood estimation
# Use method='newton' for better convergence with potential separation issues
logit_result = logit_model.fit(method='newton', maxiter=100)

# Print the detailed summary of the model
print(logit_result.summary())
```

**Model Interpretation**

The output summary table will provide the features used for building model along with coefficient of each of the feature and their p-value. The p-value in a logistic regression model is used to assess the statistical significance of each coefficient. Lesser the p-value, more significant the feature is in the model.

A positive coefficient will indicate that an increase in the value of feature would increase the odds of the event occurring. On the other hand, a negative coefficient means the opposite, i.e, an increase in the value of feature would decrease the odds of the event occurring.

Now check VIFs for presence of multicollinearity in the model.

#### **7.2.3** Evaluate VIF of features to assess multicollinearity <font color="red">[2 Marks]</font>

```python
# Import 'variance_inflation_factor'
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Make a VIF DataFrame for all the variables present
# Calculate the VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X_train_logit_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_logit_const.values, i) 
                   for i in range(X_train_logit_const.shape[1])]

# Sort the dataframe by VIF values
vif_data = vif_data.sort_values("VIF", ascending=False)

# Print the VIF values
print("VIF values for detecting multicollinearity:")
print(vif_data)

# Identify features with high VIF (typically > 5 or > 10 indicate multicollinearity)
high_vif = vif_data[vif_data["VIF"] > 5]
print("\nFeatures with high multicollinearity (VIF > 5):")
print(high_vif)
```

### **7.3 Find the Optimal Cutoff** <font color = red>[12 marks]</font>
Find the optimal cutoff to improve model performance by evaluating various cutoff values and their impact on relevant metrics.

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Define a function to plot ROC curve and find optimal threshold
def plot_roc_curve(y_true, y_prob, figsize=(10, 8)):
    """
    Plot ROC curve and calculate optimal threshold using Youden's index.
    
    Parameters:
    -----------
    y_true : array-like
        True binary labels
    y_prob : array-like
        Predicted probabilities
    figsize : tuple, optional
        Figure size (width, height)
    
    Returns:
    --------
    float
        Optimal threshold based on Youden's index
    """
    # Calculate ROC curve points
    fpr, tpr, thresholds = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.figure(figsize=figsize)
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')  # Random guess line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.grid(alpha=0.3)
    plt.show()
    
    # Find optimal threshold using Youden's index
    j_scores = tpr - fpr
    optimal_idx = j_scores.argmax()
    optimal_threshold = thresholds[optimal_idx]
    print(f"Optimal threshold based on Youden's index: {optimal_threshold:.4f}")
    print(f"At this threshold - Sensitivity: {tpr[optimal_idx]:.4f}, Specificity: {1-fpr[optimal_idx]:.4f}")
    
    return optimal_threshold
```

```python
# Call the ROC function
optimal_cutoff = plot_roc_curve(predictions_df['actual_fraud'], predictions_df['predicted_proba'])
print(f"\nOptimal cutoff value: {optimal_cutoff:.4f}")
```
